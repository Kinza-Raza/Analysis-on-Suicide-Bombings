```python
cols = ["Date","Year"]
file[cols] = file[cols].astype("int")
```

```python
from hijri_converter import convert
```

```python
hijri_date = []

for i in file[["Year","Month","Date"]].values:
    year = i[0]
    month = i[1]
    day = i[2]
    datetime_object = dt.strptime(month, "%B")
    month_number = datetime_object.month
    x = convert.Gregorian(year,int(month_number),day).to_hijri()
    h_date = x.dmyformat().split("/")[0]
    h_month = x.month_name()
    h_year = x.dmyformat().split("/")[2]
    hijri_date.append(str(h_date)+ " " + h_month + " " + str(h_year))
```

```python
file["Islamic Date"] = hijri_date
```
